```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [6,7,2,4,5]

output_notebook()

p = figure(title="simple line example", x_axis_label="x", y_axis_label="y")
p.line(x,y,legend="Temp.", line_width=2)

show(p)
```

```python
x = np.linspace(-6,6,100)
y = np.cos(x)
output_notebook()
p2 = figure(width=500, height=500)
p2.circle(x,y,size=7,color='firebrick',alpha=0.5)
show(p2)
```

```python
x1 = np.arange(0.1,3.6,0.5)
y0 = [i**2 for i in x]
y1 = [10**i for i in x]
y2 = [10**(i**2) for i in x]

output_notebook()

p3 = figure(tools = "pan,box_zoom,reset,save",
           y_axis_type="log", y_range=[0.001, 10**11], title="log axis example",
           x_axis_label="sections",y_axis_label="particles")

p3.line(x1,x1,legend="y=x")
p3.circle(x1,x1,legend="y=x", fill_color="white",size=8)
p3.line(x1,y0, legend="y=x^2",line_width=3)
p3.line(x1,y1,legend="y=10^x",line_color="red")
p3.circle(x1,y1,legend="10^x",fill_color="red",line_color="red",size=6)
p3.line(x1,y2,legend="y=10^x^2",line_color="orange",line_dash="4 4")
show(p3)
```

```python
N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii = np.random.random(size=N) * 1.5
colors = [
    "#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(50+2*x, 30+2*y)
]

# output to static HTML file (with CDN resources)
output_notebook()

TOOLS="crosshair,pan,wheel_zoom,box_zoom,reset,box_select,lasso_select,hover"

# create a new plot with the tools above, and explicit ranges
p = figure(tools=TOOLS, x_range=(0,100), y_range=(0,100))

# add a circle renderer with vectorized colors and sizes
p.circle(x,y, radius=radii, fill_color=colors, fill_alpha=0.6, line_color=None)

# show the results
show(p)
```

```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import gridplot
```

```python
N = 300
x = np.linspace(0,4.*np.pi, N)
y0 = np.sin(x)
y1 = np.cos(x)

output_notebook()

source = ColumnDataSource(data=dict(x=x,y0=y0,y1=y1))

TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select,lasso_select"

left = figure(tools = TOOLS, width = 250,height=250, title =None)
left.circle("x", "y0", source=source)

right = figure(tools=TOOLS, width=250, height=250, title=None)
right.circle("x", "y1", source=source)

p = gridplot([[left,right]])
show(p)
```
